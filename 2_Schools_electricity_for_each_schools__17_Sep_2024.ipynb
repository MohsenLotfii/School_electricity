```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import re

# List of paths for different schools
schoolPaths = [
    "/content/drive/MyDrive/ðŸ /UdS/Thesis/Projects/Alpha",
    "/content/drive/MyDrive/ðŸ /UdS/Thesis/Projects/Horizon-du-Lac",
    "/content/drive/MyDrive/ðŸ /UdS/Thesis/Projects/Marie-Soleil-Tougas",
    "/content/drive/MyDrive/ðŸ /UdS/Thesis/Projects/De_la_ClÃ©-des-Champs",
    "/content/drive/MyDrive/ðŸ /UdS/Thesis/Projects/Ã‰cole_Domaine_Vert-Nord",
    "/content/drive/MyDrive/ðŸ /UdS/Thesis/Projects/Gaston-Pilon(300209825)",
    "/content/drive/MyDrive/ðŸ /UdS/Thesis/Projects/Gaston-Pilon(300206515)",
    "/content/drive/MyDrive/ðŸ /UdS/Thesis/Projects/Gaston-Pilon(300205293)",
    "/content/drive/MyDrive/ðŸ /UdS/Thesis/Projects/Des_PersÃ©ides,pavillon_des_PrimevÃ¨res",
]

# Loop through each school path
for schoolPath in schoolPaths:
    # Get a list of all CSV files in the directory
    csv_files = [f for f in os.listdir(schoolPath) if f.endswith('.csv')]

    # Create a figure for the plot
    plt.figure(figsize=(10, 6))

    # Loop over each CSV file and plot the data
    for file in csv_files:
        full_path = os.path.join(schoolPath, file)

        # Read the CSV file into a DataFrame
        df = pd.read_csv(full_path, delimiter=';', decimal=',')

        # Convert the second column (Date and time) to datetime
        df.iloc[:, 1] = pd.to_datetime(df.iloc[:, 1], errors='coerce')

        # Drop rows with invalid or missing datetime values
        df = df.dropna(subset=[df.columns[1]])

        # Ensure there are enough columns and clean the 'Real power' column
        if df.shape[1] >= 3:
            df.iloc[:, 2] = pd.to_numeric(df.iloc[:, 2], errors='coerce')
            df = df.dropna(subset=[df.columns[2]])

            # Normalize the real power consumption for color mapping
            min_val = df.iloc[:, 2].min()
            max_val = df.iloc[:, 2].max()

            # Create a colormap from blue (low) to red (high)
            norm = plt.Normalize(min_val, max_val)
            cmap = plt.cm.get_cmap('coolwarm')  # Blue to red colormap

            # Plot the data with color intensity based on consumption
            plt.scatter(df.iloc[:, 1], df.iloc[:, 2], c=df.iloc[:, 2], cmap=cmap, norm=norm)

            # Extract the year and month from the filename using regex
            match = re.search(r'(\d{4})-(\d{2})', file)
            if match:
                year_month = f"{match.group(1)}-{match.group(2)}"
            else:
                year_month = file  # Fallback to the filename if no date found

            # Add the extracted year and month to the legend
            plt.plot([], [], label=year_month)

    # Set axis labels and title
    school_name = os.path.basename(schoolPath)
    plt.xlabel('Temps')
    plt.ylabel('kW')
    plt.title(f'Electricity Demand ({school_name})')

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Add a grid and legend
    plt.grid(True)
    plt.legend(title="Year-Month", bbox_to_anchor=(1.05, 1), loc='upper left')

    # Adjust layout to prevent label cutting
    plt.tight_layout()

    # Show the plot
    plt.show()
```
